```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
import fastcat as fc
```

# Initialize the spectrum (Spekpy)

```python
spectrum = fc.calculate_spectrum_sp(120,12)
```

```python
print(spectrum)
```

# Initialize the spectrum (MV Spectra)

#### List the MV spectra

```python
det = fc.Detector(spectrum,'CsI-784-micrometer')

fig = plt.figure()
im = det.get_plot(plt.subplot())
fig.colorbar(im,ax = plt.gca())
# plt.xlim([0,18])
# plt.ylim([0,18])
```
